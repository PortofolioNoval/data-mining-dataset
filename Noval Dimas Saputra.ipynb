```python
combined = pd.concat([su,da,ta], ignore_index=True)
```

```python
combined.columns.values
```

```python
x = combined.iloc[:,7: 79]
```

```python
x
```

```python
y = combined.iloc[:,83: 84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 43)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
dtc.fit(x_train,y_train)
```

```python
y_predic = dtc.predict(x_test)
```

```python
y_predic
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,y_predic)
```

```python
acc
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tble = plt.figure(figsize = (10, 8))
tree.plot_tree(dtc, feature_names = x.columns.values, class_names = np.array([ 'su','da','ta']), filled = True)
plt.show()
```

```python
import seaborn as yey
from sklearn import metrics
```

```python
label = np.array([ 'su','da','ta'])
```

```python
import matplotlib.pyplot as plt
```

```python
conf_matrix = metrics.confusion_matrix(y_test, y_predic)
plt.figure(figsize=(14, 14))
yey.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
